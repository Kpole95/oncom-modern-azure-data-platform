```python
%sql
SELECT
  COUNT(*) AS total_rows,
  COUNT(CurrencyKey) AS rows_with_currency_key,
  COUNT(CostCenterKey) AS rows_with_costcenter_key,
  COUNT(VatAmount) AS rows_with_vat,
  COUNT(OrderDateKey) AS rows_with_order_date_key,
  COUNT(ShipDateKey) AS rows_with_ship_date_key,
  COUNT(DeliveredDateKey) AS rows_with_delivered_date_key
FROM silver.factpurchaseorder;
```

```python
%sql
SELECT *
FROM silver.factpurchaseorder
WHERE CurrencyKey IS NULL
   OR VatAmount IS NULL
LIMIT 20;
```

```python
%sql
SELECT
  COUNT(*) AS total_rows,
  COUNT(PartyHashKey) AS hashkey_count,
  COUNT(PartyRecordId) AS recordid_count
FROM silver.dimparty;

SELECT
  COUNT(*) AS total_rows,
  COUNT(VendorHashKey) AS hashkey_count,
  COUNT(VendorRecordId) AS recordid_count
FROM silver.dimvendor;

SELECT
  COUNT(*) AS total_rows,
  COUNT(CurrencyHashKey) AS hashkey_count,
  COUNT(CurrencyRecordId) AS recordid_count
FROM silver.dimcurrency;

SELECT
  COUNT(*) AS total_rows,
  COUNT(CostCenterHashKey) AS hashkey_count,
  COUNT(CostCenterRecordId) AS recordid_count
FROM silver.dimcostcenter;
```

```python
%sql
SELECT
  MIN(Date) AS min_date,
  MAX(Date) AS max_date,
  COUNT(*) AS total_rows,
  COUNT(DateId) AS dateid_count,
  COUNT(DateKey) AS datekey_count
FROM silver.dimdate;
```
