### Define Variables

This cell defines the entity name, manifest/model name, and raw Delta Lake output path.

Only these values should change when cloning this notebook for another Purchase entity.

```python
# Define entity-specific variables.

entityName = "Currency"
manifest = "Others"
deltaLakePath = "DeltaLake/Raw/Others/" + entityName


# This defines:

# entityName = Parties
# manifest = Purchase
# deltaLakePath = DeltaLake/Raw/Purchase/Parties

# For another entity, only change:

# entityName = "VendTable"
# or:
# entityName = "PurchaseOrder"
```

### Read Entity

This cell reads the raw CDM-style entity from ADLS.

The `readEntity()` function reads the `.cdm.json` schema file, builds a StructType schema, then reads the headerless CSV folder.

```python
# Read the Parties entity from source CSV using schema from Parties.cdm.json.

entitydf = readEntity(manifest, entityName)
display(entitydf)

# This reads:

# Tables/Purchase/Parties.cdm.json
# Tables/Purchase/Parties/

# and returns a schema-aware DataFrame.
```

### Verify Entity Schema

This cell prints the schema returned by `readEntity()`.

Use this check before writing to Delta.

```python
# Verify schema before writing to Delta.

entitydf.printSchema()
```

### Write To Delta Lake

This cell writes the raw entity DataFrame to ADLS in Delta format.

This creates the raw Delta layer for the entity.

```python
# Write Parties DataFrame to raw Delta Lake path.

writeRawToDeltaLake(entitydf, deltaLakePath)
```

### Verify Delta Output Path

This cell lists the Delta output folder to confirm that Delta files were written successfully.

```python
# Verify that Delta output files exist.

display(
    dbutils.fs.ls(
        ADLS_DEV_BASE_PATH + deltaLakePath
    )
)
```

### Read Delta Output

This cell reads the newly written raw Delta files back into a DataFrame.

This confirms that the raw Delta layer is readable.

```python
# Read raw Delta output back for verification.

dfDelta = (spark.read.format("delta")
 .option("path", ADLS_DEV_BASE_PATH + deltaLakePath)
 .load())

display(dfDelta)
```
