# Raw Ingestion - CostCenter

This notebook reads the `CostCenter` entity from the source data model and writes it into the Raw layer of the Delta Lake.

The Raw layer stores source-aligned data with minimal transformation so it can be used by downstream processing layers.

## Run Shared Libraries

This cell loads the common functions and configuration values required by the notebook.

```python
%run ../Misc/sharedlibraries
```

## Define Variables

This cell defines the entity name, source manifest, and target Raw layer path.

```python
entityName = "CostCenter"
manifest = "Others"
deltaLakePath = "DeltaLake/Raw/Others/" + entityName
```

## Read Entity

This cell reads the source entity into a DataFrame using the configured manifest and entity name.

```python
entitydf = readEntity(manifest,entityName)
display(entitydf)
```

## Verify Schema

This cell prints the DataFrame schema to confirm that the source structure was read correctly.

```python
entitydf.printSchema()
```

## Write to Delta Lake

This cell writes the DataFrame to the Raw layer in Delta format.

```python
writeRawToDeltaLake(entitydf,deltaLakePath)
```

## Verify Delta Output Path

This cell lists the files written to the target Delta Lake path.

```python
display(dbutils.fs.ls(ADLS_DEV_BASE_PATH + deltaLakePath))
```

## Read Delta Output

This cell reads the Raw Delta output back into a DataFrame to confirm that the data was written successfully.

```python
dfDelta = (spark.read.format("delta")
 .option("path", ADLS_DEV_BASE_PATH + deltaLakePath)
 .load())

display(dfDelta)
```
