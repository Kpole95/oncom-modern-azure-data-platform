# Bronze Ingestion - SalesOrderLine

This notebook reads the `SalesOrderLine` entity from the Raw layer and writes it into the Bronze schema.

The Bronze layer stores source-aligned Delta tables that can be used by downstream transformation and reporting layers.

## Run Shared Libraries

This cell loads the common functions and configuration values required by the notebook.

```python
%run ../Misc/sharedlibraries
```

## Define Variables

This cell defines the entity name and the source Raw layer path used by the notebook.

```python
Entity = "SalesOrderLine"
EntityPath = f"Sales/{Entity}/"
```

## Read Raw Data

This cell reads the entity data from the Raw Delta path into a DataFrame.

```python
entitydf = readFromDeltaPath(EntityPath)
display(entitydf)
```

## Verify Schema

This cell prints the DataFrame schema to confirm that the source structure was read correctly.

```python
entitydf.printSchema()
```

## Write to Bronze Schema

This cell writes the DataFrame into the Bronze schema as a managed Delta table.

```python
saveDeltaTableToCatalog(entitydf, "Bronze", Entity)
```
